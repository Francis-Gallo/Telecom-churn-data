```python
import pandas as pd
```

```python
# Load the dataset
df = pd.read_csv('/content/telecom_churn_data.csv')

# Display the first few rows to understand the structure
df.head()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
# Get descriptive statistics
print(df.describe(include='all'))
```

```python
# Convert 'TotalCharges' to numeric, forcing errors to NaN
df['TotalCharges'] = pd.to_numeric(df['TotalCharges'], errors='coerce')

# Check for NaNs in 'TotalCharges'
print(df['TotalCharges'].isnull().sum())

# You can drop rows with NaN in 'TotalCharges' if necessary
df.dropna(subset=['TotalCharges'], inplace=True)

```

```python
# Check unique values in 'gender'
print("Unique values in 'gender':", df['gender'].unique())

# Check other categorical variables similarly
print("Unique values in 'Partner':", df['Partner'].unique())
print("Unique values in 'Dependents':", df['Dependents'].unique())
# Add more checks for other categorical variables as needed
```

```python
# Clean the 'gender' column (ensure it's consistent)
df['gender'] = df['gender'].astype(str).str.strip().str.lower()

# Map values
gender_mapping = {'male': 0, 'female': 1}
df['gender'] = df['gender'].map(gender_mapping)

# Check for NaNs after mapping
print("Number of NaN values in 'gender' after encoding:", df['gender'].isnull().sum())
```

```python
# One-hot encode selected categorical variables
categorical_columns = ['Partner', 'Dependents', 'PhoneService', 'MultipleLines',
                       'InternetService', 'OnlineSecurity', 'OnlineBackup',
                       'DeviceProtection', 'TechSupport', 'StreamingTV',
                       'StreamingMovies', 'Contract', 'PaperlessBilling',
                       'PaymentMethod']

# Apply one-hot encoding
df = pd.get_dummies(df, columns=categorical_columns, drop_first=True)

# Check the DataFrame after encoding
print(df.head())
```

```python
# Map values for 'Churn'
churn_mapping = {'No': 0, 'Yes': 1}
df['Churn'] = df['Churn'].map(churn_mapping)

# Check for NaNs after mapping
print("Number of NaN values in 'Churn' after encoding:", df['Churn'].isnull().sum())
```
